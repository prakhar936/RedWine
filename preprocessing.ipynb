```python
x = df.drop('quality',axis = 1)
y = df['quality']
x_train,x_test,y_train,y_test = tt(x,y,random_state = 42)
```

```python
x_axis = []
y_axis = []
for i in range(1,28):
    k = knn(n_neighbors = i)
    score = cross_val_score(k,x_train,y_train)
    x_axis.append(i)
    y_axis.append(score.mean())
    
plt.plot(x_axis,y_axis)
plt.show()
print(y_axis)
```

```python
algo = {
    'Logistic Regression' : lr(max_iter=10000),
    'Decision Tree' : dt(),
    'Random Forest' : rf(),
    'K Nearest Neighbors' : knn(n_neighbors=18),
    'SVM' : SVC()
}
```

```python
algo_score = []

for algo_name, al in algo.items():
    al.fit(x_train, y_train)
    y_pred = al.predict(x_test)
    score = accuracy_score(y_test,y_pred)
    algo_score.append((algo_name, score))
    
print(algo_score)
```
